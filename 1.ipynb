```python
import numpy as ny
import pandas as pd
global str
import matplotlib.pyplot as plt

# dadane masir
path=pd.read_csv(r'C:\Users\Asus\Downloads\Telegram Desktop\1.csv')

# estekhraj kardane etelaAt az excel baraye shoroe kar
a={}
champion={}
bord=path["FTR"]
home=path["HomeTeam"]
away=path["AwayTeam"]
season=path["Season"]

# fasle avalie
fasl="1995-96"

# daravordane emtiazat har bazi ke be har team cheghad emtiaz miresad
for x in range(len(path)):
    
    # bazi haye har fasl
    if season[x]==fasl:
        
        # baraye bazi hayii ke bord 2 emtiaz dasht
        if season[x] == "1995-96":
            if bord[x]== "H":
                if home[x] in a:
                    n=int(a.get(home[x]))+2
                    a.update({home[x]: n})
                else:        
                    a[home[x]]= 2
            
            elif bord[x]=="A":
                if away[x] in a:
                    n=int(a.get(away[x]))+2
                    a.update({away[x]: n})
                else:        
                    a[away[x]]= 2
         
            else:
                if home[x] in a:
                    n=int(a.get(home[x]))+1
                    a.update({home[x]: n})
                else:
                    a[home[x]]=1
                if away[x] in a:
                    n=int(a.get(away[x]))+1
                    a.update({away[x]: n})
                else:
                    a[away[x]]=1
            
            
        # baraye bazi haii ke bord 3 emtiaz dasht
        else: 
            if bord[x]== "H":
                if home[x] in a:
                    n=int(a.get(home[x]))+3
                    a.update({home[x]: n})
                else:        
                    a[home[x]]= 3
             
            elif bord[x]=="A":
                if away[x] in a:
                    n=int(a.get(away[x]))+3
                    a.update({away[x]: n})
                else:        
                    a[away[x]]= 3
         
            else:
                if home[x] in a:
                    n=int(a.get(home[x]))+1
                    a.update({home[x]: n})
                else:
                    a[home[x]]=1
                if away[x] in a:
                    n=int(a.get(away[x]))+1
                    a.update({away[x]: n})
                else:
                    a[away[x]]=1
                
    # peyda kardane ghahreman va raftab be fasle badi            
    else:
        
        # peyda kardane ghahreman 
        max=0
        st=""
        for j in a:
            if int(a[j])>max:
                max=int(a[j])
                st=j
                    
                    
        # update kardane liste ghahremana
        if st in champion:
            n=champion.get(st)+1
            champion.update({st: n})
        else:
            champion[st]=1
        
        
        # tamdide fasle jadid
        a.clear()
        joda=fasl.split("-")
        
        if joda[1]=="99":
            fasl=str(int(joda[0])+1)+"-"+"2000"
        elif int(joda[1])==2000:
            fasl=str(int(joda[0])+1)+"-"+"01"
        elif joda[1]=="01":
            fasl=str(int(joda[0])+1)+"-"+"02"
        elif joda[1]=="02":
            fasl=str(int(joda[0])+1)+"-"+"03"
        elif joda[1]=="03":
            fasl=str(int(joda[0])+1)+"-"+"04"
        elif joda[1]=="04":
            fasl=str(int(joda[0])+1)+"-"+"05"
        elif joda[1]=="05":
            fasl=str(int(joda[0])+1)+"-"+"06"
        elif joda[1]=="06":
            fasl=str(int(joda[0])+1)+"-"+"07"
        elif joda[1]=="07":
            fasl=str(int(joda[0])+1)+"-"+"08" 
        elif joda[1]=="08":
            fasl=str(int(joda[0])+1)+"-"+"09"
        else:    
            fasl=str(int(joda[0])+1)+"-"+str(int(joda[1])+1)
      
    
    
# peyda kardane ghahremane fasle 2020-21
max=0
st=""
for j in a:
    if int(a[j])>max:
        max=int(a[j])
        st=j
                                  
# update kardane liste ghahremane fasle 2020-21
if st in champion:
    n=champion.get(st)+1
    champion.update({st: n})
else:
    champion[st]=1    


# sort kardane tedade ghahremani ha
sort = sorted(champion.items(), key=lambda x: x[1], reverse=True)
sort_champ={}
for i in sort:
	 sort_champ[i[0]]=i[1]
        
# tedad ghahremani haye har team be sorate sort shode
print(sort_champ)


# keshidane nemodare mileii       
courses = list(sort_champ.keys())
values = list(sort_champ.values())
fig = plt.figure(figsize = (10, 5))
plt.bar(courses, values, color ='lightgreen',width =0.4)
plt.show()


# keshidane nemodare dayereii
courses = list(sort_champ.keys())
values = list(sort_champ.values())
plt.pie(values, labels=courses, autopct='%1.1f%%')
plt.savefig("pie.png", dpi=200)
```
